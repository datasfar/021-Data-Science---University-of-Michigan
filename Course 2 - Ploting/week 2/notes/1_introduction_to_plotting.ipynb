## Introducción a Matplotlib
Matplotlib es una biblioteca de Python utilizada para crear gráficos y visualizaciones de datos de alta calidad. Es ampliamente usada en la comunidad científica y de análisis de datos debido a su flexibilidad y amplio rango de funcionalidades.

#### Importando matplotlib
Para comenzar a utilizar matplotlib, es común importar la biblioteca junto con su módulo de scripting pyplot, que se suele renombrar como plt para facilitar su uso.

```python
import matplotlib.pyplot as plt
```

#### Backend en matplotlib
En versiones anteriores de matplotlib se utilizaba un backend específico para Jupyter llamado nbagg. Sin embargo, ahora se usa un motor backend inline, que es más sencillo y directo para la mayoría de los usuarios. Aunque no es necesario conocer los detalles del backend, es útil saber que puedes investigar y cambiarlo si es necesario.

```python
# para verificar qué backend se está utilizando, puedes usar el siguiente código:
import matplotlib as mpl
print(mpl.get_backend())
```

#### Creando Gráficos Básicos
El primer paso para crear un gráfico en matplotlib es usar la función plot. Un gráfico tiene dos ejes: el eje x (horizontal) y el eje y (vertical). A continuación, se muestra un ejemplo básico de cómo crear un gráfico.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico
plt.plot(x, y)

# Mostrar el gráfico
plt.show()
```

#### Entendiendo los Argumentos de plot
La función plot en matplotlib es muy flexible y puede aceptar cualquier número de argumentos no nombrados y argumentos nombrados. Los argumentos no nombrados se interpretan como pares x, y.

```python
plt.plot(x, y)
```

#### Añadiendo Títulos y Etiquetas
Para hacer los gráficos más informativos, se pueden añadir títulos y etiquetas a los ejes.

```python
plt.plot(x, y)
plt.title("Mi Primer Gráfico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

#### Personalizando los Gráficos
Matplotlib permite una amplia personalización de los gráficos, como cambiar los colores, estilos de línea y marcadores.

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("Gráfico Personalizado")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

#### Guardando los Gráficos
Además de mostrar los gráficos en pantalla, matplotlib permite guardarlos en archivos de imagen en diversos formatos (PNG, JPG, SVG, etc.).

```python
plt.plot(x, y)
plt.title("Gráfico Guardado")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.savefig("mi_grafico.png")
```


### (EXTRA) Diferentes maneras de trabajar con matplotlib. *Revisar sección*
En matplotlib podemos crear gráficos utilizando dos enfoques diferentes: la "capa de artista" y la "interface de scripting Pyplot".

#### Interfaz de Scripting de Pyplot
La interfaz de scripting de pyplot es la forma más sencilla de crear gráficos en matplotlib. Es similar a usar comandos en una calculadora gráfica. Pyplot gestiona muchas cosas por ti, lo que facilita la creación rápida de gráficos sin necesidad de preocuparse por los detalles subyacentes.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear un gráfico utilizando la interfaz de scripting
plt.plot(x, y)
plt.title("Gráfico con Pyplot")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

#### Capa de Artista
La capa de artista es la infraestructura subyacente en matplotlib que maneja todos los elementos visuales en un gráfico. Cada parte de un gráfico, como ejes, títulos, etiquetas, líneas, puntos, etc., es un "artista". La capa de artista te permite tener un control más detallado sobre cada uno de estos elementos.

```python
import matplotlib.pyplot as plt

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico utilizando el objeto Axes
ax.plot(x, y)

# Añadir títulos y etiquetas
ax.set_title("Gráfico con la Capa de Artista")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")

# Mostrar el gráfico
plt.show()
```

### Diferencia entre la Interfaz de Scripting y la Capa de Artista
#### Interfaz de Scripting (Pyplot):
    Simplicidad: Está diseñada para ser fácil de usar, ocultando los detalles complejos.
    Rapidez: Ideal para generar gráficos rápidamente.
    Automatización: Gestiona automáticamente la creación de figuras y ejes.
#### Capa de Artista:
    Control detallado: Permite personalizar cada elemento del gráfico.
    Flexibilidad: Puedes ajustar y manipular directamente los componentes del gráfico.
    Verbosidad: Requiere más líneas de código y una comprensión más profunda de matplotlib.

#### La Capa de Scripting de Pyplot
La capa de scripting de pyplot no es mágica; simplemente realiza algunas tareas detrás de escena por nosotros. Cuando llamamos a la función plot de pyplot, la capa de scripting verifica si ya existe una figura. Si no existe, crea una nueva figura y luego devuelve los ejes (Axes) de esta figura.

##### Acceso a la Figura y los Ejes Actuales
No necesitamos almacenar la figura creada porque siempre podemos acceder a la figura actual usando la función gcf (que significa "get current figure", obtener la figura actual). De manera similar, podemos obtener los ejes actuales usando la función gca ("get current axes", obtener los ejes actuales).

```python
# primero, creamos una nueva figura y graficamos el punto (3,2) usando un marcador de círculo ('o'):
import matplotlib.pyplot as plt

# Crear una nueva figura
plt.figure()

# Graficar el punto (3, 2) con un marcador de círculo
plt.plot(3, 2, 'o')
```

```python
# obtenemos una referencia a los ejes actuales
ax = plt.gca()

# finalmente, configuramos las propiedades de los ejes. La función axis toma cuatro parámetros: un valor mínimo 
# y máximo para x, y valores mínimos y máximos para y.
ax.axis([0, 6, 0, 10])
```

### Concepto de Artistas en matplotlib
En matplotlib, un "artista" es cualquier objeto que puede ser renderizado en una figura. Esto incluye todo, desde líneas y puntos hasta etiquetas y leyendas. Los artistas son los elementos básicos que componen una visualización en matplotlib.

#### Tipos de Artistas
    Figuras: El contenedor principal que puede contener uno o más ejes.
    Ejes: Las áreas dentro de una figura donde se dibujan los gráficos.
    Parche: Formas geométricas como círculos, rectángulos, etc.
    Texto: Cualquier tipo de texto, incluyendo títulos y etiquetas.
    Líneas: Los datos gráficos, como las líneas de un gráfico.

```python
# podemos hacer llamadas sucesivas a la función plot antes de renderizar el gráfico. Los datos se añadirán como 
# diferentes series y se colorearán de manera diferente.
plt.figure()

# Graficar varios puntos
plt.plot(1.5, 1.5, 'o')
plt.plot(2, 2, 'o')
plt.plot(2.5, 2.5, 'o')
```

### Obteniendo los Hijos de un Objeto de Ejes
Podemos ir más allá con el objeto Axes, hasta el punto de obtener todos los objetos hijos que contiene. Usamos la función get_children del objeto de ejes para hacer esto.

```python
# Obtener los ejes actuales - esto creará una nueva figura ya que estamos en una nueva celda
ax = plt.gca()

# Añadir un punto
plt.plot(1.5, 1.5, 'o')

# Obtener y mostrar los hijos del objeto de ejes
children = ax.get_children()
print(children)
```
