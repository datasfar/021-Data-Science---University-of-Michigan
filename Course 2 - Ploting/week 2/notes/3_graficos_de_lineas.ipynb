### Gráficos de Lineas
Los gráficos de lineas son representaciones gráficas donde los datos se muestran como puntos conectados por líneas rectas, lo que facilita la visualización de tendencias y patrones en los datos a lo largo del tiempo o de otra variable continua.

```python
import matplotlib.pyplot as plt
import numpy as np

# creamos los datos (array de valores) y array con los cuadrados de los valores
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2

plt.figure()
#  se grafica con el estilo '-o', lo que significa que se usarán líneas rectas para conectar los puntos y 
# los puntos serán marcados con un círculo (o).
plt.plot(linear_data, '-o', exponential_data, '-o')
```

```python
plt.plot(linear_data, '-o', exponential_data, '-o')
# podemos usar "--r" para hacer una line dasheada 
plt.plot([22,44,55], '--r')

# esta vez no hemos llamado a figure, por defecto la interface de scripting de pyplot llama a plt.gcf()
# get_current_figure() lo que devuelve la figura actual
```

```python
# creamos las labels y el título
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')

# añadimos la leyenda con las entradas que que no habiamos especificado labels en los plots
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.plot([22,44,55], '--r')
plt.legend(['Baseline', 'Competition', 'Us'])
```

```python
# añadimos labels, titulo y leyenda 
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.legend(['Baseline', 'Competition'])

# gca() devuelve el objeto actual
# fill_between se usa para rellenar el área entre dos conjuntos de datos
# facecolor='blue' especifica el color del relleno
# alpha=0.2 establece la transparencia del relleno. Un valor de 0 es completamente transparente y 1 es completamente opaco
plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, exponential_data, 
                       facecolor='blue', 
                       alpha=0.25)
```

#### Trabajar con fechas

```python
# generamos un rango de fechas
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

# reploteamos nuestros datos usando el rango de fechas, antes solo teniamos 1 valor para x e y por lo 
# que este se duplicaba, ahora utiliza las fechas como x
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')
```

```python
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

# obtenemos la variable que apunta al eje x
x = plt.gca().xaxis

# iteramos sobre las labels del eje x y las rotamos para una mejor lectura
for item in x.get_ticklabels():
    item.set_rotation(45)

# cambiamos nuestras labels
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')

# matplotlib permite el uso de LaTeX indicando "$" delante de la función
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance")
```
