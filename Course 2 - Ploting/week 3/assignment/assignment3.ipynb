In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import get_cmap
```

```python
# Generamos los datos
np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
# transposicionamos el dataframe para tener los años como columnas
df = df.T

# usamos describe para tener una visión general de como se ven los datos 
df.describe()
```

```python
# cambiar los nombres de las columnas a tipo string
df.columns = df.columns.astype(str)

# calcular medias y desviaciones estándar para cada año
means = df.mean()
std_devs = df.std()

# calcular intervalos de confianza del 95%, se utiliza el factor 1.96 que corresponde al valor 
# crítico de la distribución normal para un intervalo de confianza del 95%.
confidence_intervals = 1.96 * (std_devs / np.sqrt(len(df)))
```

```python
# para visualizar estos intervalos en el gráfico, se normalizan las medias con respecto a un
# valor de referencia (y = 40000)

# la normalización transforma las medias ajustadas a una escala que va de -1.96 a 1.96 
# (esto está relacionado con el intervalo de confianza del 95%).
norm = Normalize(vmin=-1.96, vmax=1.96)

# obtiene un colormap que muestra una transición de colores entre dos extremos
cmap = get_cmap('PiYG')

# Valor de referencia para la línea horizontal
y = 40000
```

```python
# Crear un DataFrame para almacenar los valores normalizados y colores
colors = pd.DataFrame(index=range(4), columns=['Value', 'Color'])
colors['Value'] = norm((means - y) / std_devs)
colors['Color'] = [cmap(value) for value in colors['Value']]
```

```python
# Configuración del gráfico
fig, ax = plt.subplots(figsize=(12, 8))

# Crear barras
bars = ax.bar(range(4), means, yerr=confidence_intervals, color=colors['Color'])

# Añadir línea horizontal de referencia
ax.axhline(y=y, color='black', alpha=0.3)
ax.text(3.65, y, f'y={y}', verticalalignment='center')

# Etiquetas del eje x y configuración
ax.set_xticks(range(4))
ax.set_xticklabels(['1992', '1993', '1994', '1995'])
ax.set_xlabel('Year')
ax.set_ylabel('Mean Votes')
ax.set_title('Average Votes per Year')

# Añadir colorbar como leyenda
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='horizontal')
cbar.set_label('Normalized Deviation from 40000')
cbar.set_ticks([-1.96, 0, 1.96])
cbar.set_ticklabels(['< -1.96', '0', '> 1.96'])

# Mostrar gráfico
plt.show()
```

#### Interpretación del Gráfico

 - Barras: Las alturas de las barras muestran las medias de votos para cada año.
 - Barras de Error: Muestran el intervalo de confianza del 95% alrededor de cada media, dando una idea de la precisión de la estimación de la media.
 - Colores de las Barras: La variación de color muestra la desviación normalizada de cada media con respecto al valor de referencia 40000. Colores cercanos al rojo indican valores más bajos que el de referencia, mientras que colores cercanos al azul indican valores más altos.
 - Línea Horizontal: La línea en 40000 proporciona un contexto de referencia para comparar las medias de los años.
 
Este gráfico es útil para ver cómo se comparan las medias anuales con respecto a un valor de referencia específico y para observar la variabilidad de las medias a lo largo de los años.
